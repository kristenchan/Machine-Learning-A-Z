## Importing the Dataset 讀入資料

```r
path = '/Users/sun/Desktop/Simple_Linear_Regression/'
dataset = read.csv(file.path(path,'Salary_Data.csv'))
```

# Splitting the dataset into the Training set and Test set

+ Training set : 用來建立 Model
+ Test set : 拿來跟 Model 預測出來的值做驗證，以確定 Model 的好壞

```r
library(caTools)
set.seed(123)
split = sample.split(dataset$Salary, SplitRatio = 2/3 )
training_set = subset(dataset, split == TRUE)
test_set = subset(dataset, split == FALSE)
```

> [Note] <br>
> 1. 雖然 0.75 / 0.25 是很好的切分比例，但因為這組資料剛好是 30 筆，所以拿 1/3 去當 Test set , 2/3 去當 Training set

## Fitting Simple Linear Regression to the Training set

拿 Training set 建立 Simple Linear Regression Model

```r
regressor = lm(formula = Salary ~ YearsExperience,
               data = training_set)
```

```r
summary(regressor)
```

> [Note]
> 1. 利用 `summary` 可以得知更多 Model 的相關訊息
> 2. 在 `Coefficients` 中，Pr(>|t|)表示 P-Value。<br>
     即對迴歸係數進行統計檢定的 P-Value ，可以用來看該迴歸係數對 Dependent Variable 的結果是否顯著
     + `***` : 有高度顯著關係
     + NO `*` : 沒有顯著關係
     + 越小的 P-Value 表示越顯著，通常 P-Value 小於 0.05 就表示有顯著關係
> 3. Multiple R-squared : 表示 Model 的解釋能力     

## Predicting the Test set results

```r
y_pred = predict(regressor, newdata = test_set)
```

## Visualising the results

#### The Training set 

```r
library(ggplot2)
ggplot() +
  geom_point(aes(x = training_set$YearsExperience, y = training_set$Salary),
             colour = 'red') +
  geom_line(aes(x = training_set$YearsExperience, y = predict(regressor, newdata = training_set)),
            colour = 'blue') +
  ggtitle('Salary vs Experience (Training set)') +
  xlab('Years of experience') +
  ylab('Salary')
```

#### The Test set 

```r
library(ggplot2)
ggplot() +
  geom_point(aes(x = test_set$YearsExperience, y = test_set$Salary),
             colour = 'red') +
  geom_line(aes(x = training_set$YearsExperience, y = predict(regressor, newdata = training_set)),
            colour = 'blue') +
  ggtitle('Salary vs Experience (Test set)') +
  xlab('Years of experience') +
  ylab('Salary')
```
