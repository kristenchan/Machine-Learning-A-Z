## The Kernel Trick

Gaussian RBF Kernel
$$
K(\overrightarrow{x},\overrightarrow{l^i}) = e^{-\frac{{\lVert \overrightarrow{x}-\overrightarrow{l^i} \rVert}^2}{2\sigma^2}}
$$

![](plot_3_4_6.png)

+ 紅色的點：距離中心點較遠，因此 ${\lVert \overrightarrow{x}-\overrightarrow{l^i} \rVert}^2$ 很大，導致 $e^{-很大} \, 
\to 0$
+ 綠色的點：距離中心點很近，因此 ${\lVert \overrightarrow{x}-\overrightarrow{l^i} \rVert}^2$ 很小，導致 $e^{-很小} \, 
\to 1$

![](plot_3_4_7.png)

![](plot_3_4_8.png)

> [Note] <br>
> 中間的紅點視為 Gaussian 的中心，綠點因為離中心比較近，會導致 $e^{-很小} \, \to 1$ 表示 Gaussian 中非深藍色的部分， 而外圍的紅點因為離中心較遠，會導致 $e^{-很大} \, \to 0$ 表示為  Gaussian 中深藍色的部分

![](plot_3_4_9.png)

> [Note] <br>
> $\sigma$ 的大小，會影響切分的範圍

![](plot_3_4_10.png)

>[Note] <br>
>如果是較複雜的形式，可以結合兩個 Gaussian

## Splitting the dataset into the Training set and Test set 切分訓練及測試樣本

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

## Feature Scaling 數值型資料尺度轉換

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Fitting Kernel(RBF) SVM to the Training set

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0)
classifier.fit(X_train, y_train)
```

> [Note] <br>
> 因為這組資料不是線性可分的，因此可以嘗試使用 kernel=rbf

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

## Visualising the Training set results

```python
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Kernel SVM (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

## Visualising the Test set results

```python
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Kernel SVM (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
