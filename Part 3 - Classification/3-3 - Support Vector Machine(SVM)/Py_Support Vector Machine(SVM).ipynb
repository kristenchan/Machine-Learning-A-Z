## Importing the Libraries 載入套件

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

## Importing the Dataset 讀入資料

利用 pd.read_csv 來載入資料，這組資料是一個 400 列(row) 5 行(column) Social Network的資料，其中 Purchased 表示有沒有購買。

```python
path = '/Users/hsinyu/Desktop/Support_Vector_Machine(SVM)/'
dataset = pd.read_csv( path+'Social_Network_Ads.csv' )
```

## Dependent & independent variables 定義解釋變數及反應變數

```python
X = dataset.iloc[:, [2, 3]].values #Age,EstimatedSalary
y = dataset.iloc[:, 4].values #Purchased
```

## SVM Intuition

SVM 是一個同時可以運用在線性及非線性資料的分類演算法。
<br>
SVM 主要目標就是在一個由不同類別混合而成的資料集中，找出一個超平面(hyperplane)，將不同的類別分得越開越好
> [Note] <br>
> 使用超平面一詞的原因是，因為它會將原始資料轉換到更高的維度 (dimension)，而在這個高維度空間找到一個分開不同類別的平面，而此平面稱之為超平面

![](plot_3_3_1.png)

#### Support Vector Machine(SVM) ：
利用在訓練資料集 (training data set)中 support vector (最靠近超平面的兩個點，雖然是個點但也可以視為是一個向量<黑色粗線>) 找到超平面(hyperplane)來分類資料。而這個 hyperplane 會是具有最大 margin 的 hyperplane ，也就是 maximum marginal hyperplane。要找出 maximum marginal hyperplane 的原因是，因為其具有較高的分類準確性。

## Splitting the dataset into the Training set and Test set 切分訓練及測試樣本

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

## Feature Scaling 數值型資料尺度轉換

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Fitting SVM to the Training set

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)
```

>[Note] <br>
>可以依據資料型態選擇適合的 kernel

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

## Visualising the Training set results

```python
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('SVM (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

>[Note] <br>
>剛剛在建立 Model 時選擇 linear kernel ，因為資料是二維的所以 Decision Boundary 是一條線
